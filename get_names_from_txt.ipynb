```python
import stanza
import sys
import os
import re
import numpy as np
```

```python
##load stanza model
language = 'en'
stanza.download(language, verbose = False)
nlp = stanza.Pipeline(language, processors='tokenize, mwt, ner', verbose = False)
```

```python
def extract_words(text):
    """Finds all words in a text string"""
    words = re.findall(r"[\w']+", text)

    words = list(map(lambda w: w.upper(), words))
    return words

def extract_characters(text):
    """Finds all characters in a text string"""
    characters = re.findall(r"[a-z A-Z]", text)
    
    characters = list(map(lambda c: c.upper(), characters))
    return characters
```
